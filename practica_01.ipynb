```python
# Instalación de todas las librerías a utilizar 
from pylab import *
from scipy import signal
import time
import math 
import matplotlib.pyplot as plt
import numpy as np
import random as rd
import logging
# Esta es la librería encargada de realizar el Multithreading 
import concurrent.futures
import threading
```

```python
logging.basicConfig(level = logging.INFO, format = '[%(levelname)s] (%(threadName)-s) %(message)s')
```

```python
# Función que calcula el seno 
def seno(valor):
  result = math.sin(valor)
  return print(str(result))

# Función que calcula el coseno 
def coseno(valor):
  result = math.cos(valor)
  return print(str(result))

# Función que devuelve una señal cuadrada
def cuadrada(valor):
  result = rd.choice([0, 2])
  return print(str(result))

# Función que calcula la dientes de sierra 
def sierra(valor):
  result = signal.sawtooth(valor) 
  return print(str(result))
```

```python
# Función que salva los datos en un archivo.txt
def guardar(dato1, dato2, dato3, dato4):
  guarda = []
  valores = [dato1, dato2, dato3, dato4]
  guarda.extend(valores)
  print(guarda)
  archivo = open("Datos.txt", "a+")
  time.sleep(5)
  archivo.write(str(guarda) + "\n")
  print(archivo)
  archivo.close()
```

```python
numeroDeIteraciones = range(5)
threads = []
results = []
for i in numeroDeIteraciones:
    t1 = threading.Thread(name = 'hilo-1', target=seno, args=(i, ))
    t2 = threading.Thread(name = 'hilo-2', target=coseno, args=(i, ))
    t3 = threading.Thread(name = 'hilo-3', target=cuadrada, args=(i, ))
    t4 = threading.Thread(name = 'hilo-4', target=sierra, args=(i, ))
    threads.append(t1)
    threads.append(t2)
    threads.append(t3)
    threads.append(t4)

for thread in threads:
    thread.start()
    t = thread.join()
    results = np.append(results, t)


print(results)
#####################################################################################################################
# # Aquí se crea el uso de los threads 
# with concurrent.futures.ThreadPoolExecutor() as executor:
#     # Este es un arreglo, que guardará el resultado de cada executor
#     futures = []
#     for i in numeroDeIteraciones:
#         # Aquí se agrega el resultado del executor por cada iteración en el arreglo futures
#         futures.append(executor.submit(seno, i))
#         futures.append(executor.submit(coseno, i))
#         futures.append(executor.submit(cuadrada, i))
#         futures.append(executor.submit(sierra, i))
#         #futures.append(executor.submit(guardar, data1, data2, data3, data4))
#     for future in concurrent.futures.as_completed(futures):
#         # Aquí se recorre el arreglo cuando todas las tareas son completadas, para proceder a imprimir el resultado
#         data = future.result()
#         print(data)
#         time.sleep(5)
#####################################################################################################################
# if __name__=='__main__':
#     numeroDeIteraciones = 5
#     iniThreads=[]
#     threads=[]
#     for ni in range(numeroDeIteraciones):
#         t1 = Thread(target=seno, args= (ni))
#         t2 = Thread(target=coseno, args=(ni))
#         t3 = Thread(target=cuadrada, args=(ni))
#         t4 = Thread(target=sierra, args=(ni))
#         t1.start()
#         t2.start()
#         t3.start()
#         t4.start()
#         threads.extend([t1, t2, t3, t4])
#     for thread in threads:
#         thread.join()
#####################################################################################################################
# iwth concurrent.futures.ThreadPoolExecutor() as executor:
#     results0 = [executor.submit(seno, i) for i in range(numeroDeIteraciones)]
#     results1 = [executor.submit(coseno, i) for i in range(numeroDeIteraciones)]
#     results2 = [executor.submit(cuadrada, i) for i in range(numeroDeIteraciones)]
#     results3 = [executor.submit(sierra, i) for i in range(numeroDeIteraciones)]
#     for f in concurrent.futures.as_completed([results0, results1, results2, results3]):
#         print(f.results())
####################################################################################################################
# threads = []
# for _ in range(numeroDeIteraciones):
#     t = threading.Thread(target=seno, args=[1])
#     t.start()
#     threads.append(t)

# for i in range(numeroDeIteraciones):
#     t1 = threading.Thread(target=seno, args= )
#     t2 = threading.Thread(target=coseno, args= 5)
#     t3 = threading.Thread(target=cuadrada)
#     t4 = threading.Thread(target=sierra, args= 5)
#     t1.start()
#     t2.start()
#     t3.start()
#     t4.start()
#     threads.append(t1)
#     threads.append(t2)
#     threads.append(t3)
#     threads.append(t4)

# for thread in threads:
#     thread.join() 
####################################################################################################################
```
